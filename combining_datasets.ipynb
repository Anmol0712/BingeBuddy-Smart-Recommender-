```python
import ast
def extract_year_safe(value):
    if pd.isna(value) or value == "Unknown":
        return None
    try:
        data = ast.literal_eval(value)
        from_date = data.get("from")
        if from_date:
            return from_date[:4]  
    except:
        return None
    return None
```

```python
df1["release_year"] = df1["aired_date"].apply(extract_year_safe)
```

```python
df1["release_year"] = df1["release_year"].astype(float).astype('Int64')
```
